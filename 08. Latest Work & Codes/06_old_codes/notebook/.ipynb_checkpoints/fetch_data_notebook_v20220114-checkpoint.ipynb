```python
# team name
team = "Fiorentina"

# team match ids
match_ids = match_df.loc[
    (match_df["home_team_name"] == team) |
    (match_df["away_team_name"] == team), "match_id"
].values
```

```python
event_df = pd.DataFrame()

for match_id in tqdm(match_ids, total=len(match_ids)):
    event = make_event_df(match_id)

    event_df = pd.concat([event_df, event])

event_df = event_df.reset_index(drop=True)
```

```python
all_passes = event_df.loc[
    (event_df["team_name"] == "Fiorentina") &
    (event_df["type_displayName"] == "Pass") 
    # (event_df["outcomeType_displayName"] == "Successful")
].copy()
```

```python
def progressive_passes(x):
    start_x, end_x = x

    # from wyscout glossary
    # if start_x <= 50.0 and end_x <= 50.0:
    #     # team's half
    #     if end_x - start_x >= 30.0:
    #         return True
        
    # if start_x <= 50.0 and end_x > 50.0:
    #     # team's half --> opponent's half
    #     if end_x - start_x >= 15:
    #         return True

    if start_x > 50.0 and end_x > 50.0:
        # opponent's half
        if end_x - start_x >= 10:
            return True
    
    return False
```

```python
all_passes.shape
```

```python
prog_passes = all_passes[
    all_passes[['x', "endX"]].apply(
        lambda x: progressive_passes(x), axis=1
    )
]
```

```python
temp_df = event_df.loc[
    prog_passes.index + 1
]

temp_df = temp_df.loc[
    temp_df["player_name"] == "Dusan Vlahovic"
]
```

```python
temp_df[['x', 'y', 'endX', 'endY']]
```

```python
prog_passes = prog_passes.loc[temp_df.index - 1].copy()
```

```python
def not_throw_ins(x):
    for qualifier in x:
        if qualifier["type"]["value"] == 107:
            return False
    return True

prog_passes = prog_passes[prog_passes["qualifiers"].apply(
    lambda x: not_throw_ins(x)
)]
```

```python
from selenium import webdriver
import json

options = webdriver.ChromeOptions()
options.add_experimental_option('excludeSwitches', ['enable-logging'])

# ready the chrome-drivers
driver = webdriver.Chrome("C:/Chrome_Drivers/chromedriver", options=options)

for i in range(1, 20000):
    link = f"https://understat.com/match/{str(i)}"
    
    try:
        driver.get(link)

        # get the match data
        content = driver.execute_script("return shotsData;")
        
        xg_file = open(f"{str(i)}.json", "w")
        json.dump(content, xg_file, indent=2)
        xg_file.close()
        
    except Exception as e:
        print(f"Match ID {i} not present")

# close the driver
driver.close()
```
